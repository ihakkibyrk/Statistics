## Correlation & Covariance

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
tv_hours = [3,5,2,0,5,3,1,4,3,4]
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6]
```

```python
stats.pearsonr(tv_hours, GPA) # Corr coefficient and Pi value
```

```python
stats.spearmanr(tv_hours, GPA)
```

```python
tau, pvalue = stats.kendalltau(tv_hours, GPA) 
```

```python
pvalue
```

## Linear Regression

```python
slope, intercept, rvalue, pvalue, stderr = stats.linregress(tv_hours, GPA)
```

```python
# R_squared = r_value**2
```

```python
x_p = np.array([0, 1, 2, 3, 4, 5])
y_p = intercept + slope * x_p
```

```python
x_p
```

```python
y_p
```

```python
x_new = np.linspace(0, 5, 100)
y_new = intercept + slope * x_new
```

```python
x_new
```

```python
y_new
```

```python
plt.figure(figsize = (8,6))
ax = plt.axes()
ax.scatter(tv_hours, GPA)
ax.plot(x_new, y_new)
ax.set_xlabel('tv_hours')
ax.set_ylabel('GPA')

```
